```python
%%time

test = {
    'words':
    """
        Lorem ipsum dolor sit amet, consectetur adipiscing 
        elit. Mauris adipiscing adipiscing placerat. 
        Vestibulum augue augue, 
        pellentesque quis sollicitudin id, adipiscing.
        """,
    'list':
    list(range(10000)),
    'dict':
    dict((str(i), 'a') for i in range(10000)),
    'int':
    100,
    'float':
    100.123456
}


def sizeof_fmt(num, suffix='B'):
    for unit in ['', 'Ki', 'Mi', 'Gi', 'Ti', 'Pi', 'Ei', 'Zi']:
        if abs(num) < 1024.0:
            return "%3.1f %s%s" % (num, unit, suffix)
        num /= 1024.0
    return "%.1f %s%s" % (num, 'Yi', suffix)


def get_file_size(filename):
    statinfo = os.stat(filename)
    return sizeof_fmt(statinfo.st_size)


def save_pickle(item, filename):
    with open(filename, 'wb') as file:
        pickle.dump(item, file)
    print('\n'
          'SAVED  > ', filename, ' | ',
          get_file_size(filename), ' | ', 'length: ', len(item), '\n')
    return True


def load_pickle(filename):
    with open(filename, 'rb') as file:
        obj = pickle.load(file)
        print('\n'
              'LOADED > ', filename, ' | ',
              get_file_size(filename), ' | ', 'length: ', len(obj), '\n')
        return obj


save_pickle(test, 'test.pickle')

len(load_pickle('test.pickle'))
```

```python
ax = all_players_career_winshare.plot('hist', ylim=(0, 2000), xlim=(-20, 300), bins=32, figsize=(10,5))

def annotate_ax(ax, player_name, height):
    label = player_name + ' (' + str(all_players_career_winshare[player_name]) + ')'
    ax.annotate(label,
                xycoords='data', # xy coord mode
                xy=(all_players_career_winshare[player_name], 0), # xy coord value
                textcoords='offset points', # text coord mode (in this case offset the text)
                xytext=(0, height), # text coord value
                rotation=45,
                verticalalignment='left',
                horizontalalignment='bottom')

annotate_ax(ax, 'kareem abdul-jabbar', 100)
annotate_ax(ax, 'michael jordan', 100)
annotate_ax(ax, 'larry bird', 100)
annotate_ax(ax, 'magic johnson', 100)
annotate_ax(ax, 'kobe bryant', 100)
annotate_ax(ax, 'mike bibby', 100)

annotate_ax(ax, 'lebron james', 10)
annotate_ax(ax, 'kevin durant', 10)
annotate_ax(ax, 'james harden', 10)
annotate_ax(ax, 'stephen curry', 10)
annotate_ax(ax, 'kawhi leonard', 10)
annotate_ax(ax, 'mike conley', 10)

    
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

ax.set_xlabel('Career Win Shares')
ax.set_ylabel('Number of Players')

ax.set_xticks(range(0, 300, 25))
```

```python
def annotate_ax(ax, data, player_name, height):
    label = player_name + ' (' + str(all_players_career_winshare[player_name]) + ')'
    ax.annotate(label,
                xycoords='data', # xy coord mode
                xy=(data[player_name], 0), # xy coord value
                textcoords='offset points', # text coord mode (in this case offset the text)
                xytext=(0, height), # text coord value
                rotation=45,
                verticalalignment='left',
                horizontalalignment='bottom')

def plot_player_stats(
    stat_type='totals', stat='PTS', year='Career', title='Histogram: NBA Career PTS', annotation_players=None):
    
    data = all_players_stats[stat_type].xs(year, level=1, drop_level=True)[stat]
    
    ax = data.plot('hist', figsize=(10,5))
    ax.set_title(title)
    ax.set_xlabel(str(year) + ' ' + stat)
    ax.set_ylabel('Number of Players')
    
    # Hide the right and top spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    # Only show ticks on the left and bottom spines
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    
    for (player, height) in annotation_players:
        annotate_ax(ax, data, player, height)

annotation_players = [
    ('kobe bryant', 100), 
    ('michael jordan', 100), 
    ('yao ming', 100),
    ('kareem abdul-jabbar', 100),
    ('magic johnson', 100),
    ('larry bird', 100),
]
plot_player_stats(annotation_players=annotation_players)
```
